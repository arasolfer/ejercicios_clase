### **Arreglar grafico**

Arreglar el código de pandas y matplotlib

```python
# librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# datos de ejemplo
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""
print('Tipo de datos: ', type(data))
print(data)
```

```python
# importar stringIO
from io import StringIO

# simular que es csv
csv_data = StringIO(data)

# convertir a dataframe
df = pd.read_csv(csv_data)
df.head()
```

```python
# tipos de datos
print(df.dtypes)
```

```python
# fecha a datetime
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
# tipos de datos
print(df.dtypes)
```

```python
# tamaño de la figura
plt.figure(figsize=(10, 6))

# grafico de lineas de las ventas
plt.plot(df['fecha'], df['ventas'], color='skyblue',linewidth=2, marker='o')
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.show()
```

```python
# tamaño de la figura
plt.figure(figsize=(10, 6))

# gráfico de dispersión
plt.scatter(df['fecha'], df['ventas'], )
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid()
plt.show()
```

```python
# tamaño de la figura
plt.figure(figsize=(10, 6))

# histograma de las ventas
plt.hist(df['ventas'], bins=5, color='green', edgecolor='white', alpha=0.75)
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.show()
```
