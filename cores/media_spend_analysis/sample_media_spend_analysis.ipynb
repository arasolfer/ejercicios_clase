### **An√°lisis de media spend dataset**

```python
# importacion de las librerias necesarias 
import numpy as np
import pandas as pd
```

```python
# importacion de los datos y conversion a dataframe
media = pd.read_csv('../data/sample_media_spend_data.csv')
print(media.head(10))
```

```python
# ultimas filas del dataset
print(media.tail(5))
```

```python
# informacion del dataframe
print(media.info())
```

```python
# estadisticas descriptivas basicas del dataframe
print("Estadistica descriptiva del dataset")
print(media.describe())
```

```python
# descripcion de los tipos de datos en columnas
print(media.dtypes)
```

```python
# correccion de los tipos de datos objetos a cadenas
media = media.astype({col: 'string' for col in media.select_dtypes(include='object').columns})
print(media.dtypes)
```

```python
# correccion del dato tipo cadena calendar week a datetime
media['Calendar_Week'] = pd.to_datetime(media['Calendar_Week'])
print(media.dtypes)
```

```python
# visualizacion de los datos corregidos
media.head(10)
```

```python
# identificando los valores nulos en el dataframe
# cantidad de datos completos, sin NA, es decir, el total de datos
qsna = media.shape[0]-media.isnull().sum(axis=0)
# cantidad de datos que si son na
qna=media.isnull().sum(axis=0)
# proporcion de datos con na y sin na 
ppna=round(100*(media.isnull().sum(axis=0)/media.shape[0]),2)
# constuir la tabla
aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
na=pd.DataFrame(data=aux)
# ordenar los valores de la tabla
na.sort_values(by='Na en %',ascending=False)
```

```python
# mostrar solo valores maximos de las columnas 
print(media.max())
```

```python
# mostrar solo valores minimos de las columnas 
print(media.min())
```

```python
# mostra paid_views mayor a 2699
media.Paid_Views>=2699
# devuelve datos tipo bool
```

```python
# se realiza un filtrado y drop la columna calendar week
media_filter = media.drop(columns=['Calendar_Week'],axis = 1 , inplace = False)
media_filter.head()
```

```python
# mostrar paid_views que sean mayores a 2699 o organic_views mayores a 4110
media_filter = media[(media.Paid_Views>=2699) | (media.Organic_Views>=4110)]
print(media_filter.shape)
media_filter.head()
```

```python
# tambien se puede hacer 
media_filter = media.query('Paid_Views>=2699 or Organic_Views>=4110')
print(media_filter.shape)
media_filter.head()
```

```python
# ordenar el dataset por paid_views
media_ordenado = media.sort_values(by='Paid_Views',ascending=False)
media_ordenado.head()
```

```python
# setear division como indice
media_filter = media.set_index('Division',inplace = False)
media_filter.head()
```

```python
# identificar todas las divisiones existentes
print(len(media['Division'].unique()))
media['Division'].unique()
```

```python
# agrupar datos por division
media_grouped = media.groupby('Division')
media_grouped.head()
```

```python
# sumar todos los datos de paid_views por division
media_grouped['Paid_Views'].sum()
```

```python
# ventas por categoria
media_grouped['Sales'].sum()
```

```python
# promedio todos los datos por division
media_grouped['Paid_Views'].mean()
```

```python
# identificar el valor minimo y maximo de sales 
max_sales = media['Sales'].max()
min_sales = media['Sales'].min()
```

```python
# duplicar el dataframe y agregar la columna
media_added = media
# normalizar las ventas 
media_added['Normalized']=media_added['Sales'].apply(
    lambda x: (x - min_sales)/(max_sales - min_sales)
)
media_added.head()
```
